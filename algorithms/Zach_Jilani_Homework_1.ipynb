````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Identity
#g(x) = x

x = np.arange(-5, 5, 0.01)
y = x
plt.plot(x,y)
plt.title("Identity")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Binary Step

def bin_step(x):
  return np.heaviside(x, 1)

x = np.arange(-5, 5, 0.01)
plt.plot(x, bin_step(x))
plt.title("Binary Step")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Sigmoid

def sig(x):
  return 1/(1 + np.exp(-x))
  
x = np.arange(-5, 5, 0.01)
plt.plot(x, sig(x))
plt.title("Sigmoid")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Hyperbolic Tangent(tanh)

def tanh(x):
  return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

x = np.arange(-5, 5, 0.01)
plt.plot(x, tanh(x))
plt.title("Hyperbolic Tangent")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Rectified Linear Unit (ReLU)
def relu(x):
  y = []
  for i in x:
    if i < 0:
      y.append(0)
    else:
      y.append(i)
  return y
  

x = np.arange(-5, 5, 0.01)
plt.plot(x, relu(x))
plt.title("Rectified Linear Unit (ReLU)")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
import math
from matplotlib import pyplot as plt

#Activation Function: Gaussian Error Linear Unit (GeLU)

def gelu(x):
  #Here, math must be imported since it has a built in
  #Gauss Error fn, erf. (if numpy has one I couldnt find it
  #so sorry in advance if it does)
  cdf = 0.5 * (1.0 + math.erf(x/np.sqrt(2.0)))
  return x * cdf

#this must be done since x must be a scalar
#rather than an array-like object.
GeLU = np.vectorize(gelu)

x = np.arange(-5, 5, 0.01)
plt.plot(x, GeLU(x))
plt.title("Gaussian Error Linear Unit (GeLU)")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Softplus

def soft(x):
  return np.log(1 + np.exp(x))

x = np.arange(-5, 5, 0.01)
plt.plot(x, soft(x))
plt.title("Softplus")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Leaky ReLU

def relu(x, a):
  return max(a * x, x)

LeakyReLU = np.vectorize(relu)
x = np.arange(-5, 5, 0.01)
plt.plot(x, LeakyReLU(x, .1))
plt.title("Leaky ReLU")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Scaled Exponential Linear Unit (SELU)

def selu(x):
  l = 1.0507
  a = 1.6732
  if x >= 0:
    return x * l
  else:
    return l * a * (np.exp(x) - 1)

SELU = np.vectorize(selu)

x = np.arange(-5, 5, 0.01)
plt.plot(x, SELU(x))
plt.title("Scaled Exponential Linear Unit (SELU)")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Exponential Linear Unit (ELU)

def elu(x, a):
  if x >= 0:
    return x
  else:
    return a*(np.exp(x) - 1)

ELU = np.vectorize(elu)
x = np.arange(-5, 5, 0.01)
plt.plot(x, ELU(x, 1))
plt.title("Exponential Linear Unit (ELU)")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Parametric Rectified Linear Unit (ReLU)
def prelu(x, a):
  return x if x >= 0 else a * x

PReLU = np.vectorize(prelu)
x = np.arange(-5, 5, 0.01)
plt.plot(x, PReLU(x, .2))
plt.title("Parametric Rectified Linear Unit (ReLU)")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Sigmoid Linear Unit (SiLU)

def sig(x):
  return x/(1 + np.exp(-x))
  
x = np.arange(-5, 5, 0.01)
plt.plot(x, sig(x))
plt.title("Sigmoid Linear Unit (SiLU)")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````

````python
import numpy as np
from matplotlib import pyplot as plt

#Activation Function: Gaussian

def g(x):
  return np.exp(-x**2)

x = np.arange(-5, 5, 0.01)
plt.plot(x, g(x))
plt.title("Gaussian")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid()
plt.show()
````
